## Hyperparameter

| **Hyperparameter** | **Typical value** |
| --- | --- |
| Input layer shape | Ba tavajoh be tedade vurudi hamoon oon ro moshakhas mikonim (tedade feature haye vurudi) |
| Hidden layer(s) | Mitoone beyne yek ta har chandta ke bekhaim bashe. |
| Neurons per hidden layer | Mitoone harcheghad bashe. |
| Output layer shape | Ba tavajoh be inke che masalei ro mikhaim hal konim, masalan inja chon yek Y khuruji darim 1 hastesh, ama tooye classifcation mibinim ke fargh mikone. |
| Hidden activation | Mamoolan ReLU |
| Output activation | ReLU, Linear, Softmax, etc. |
| Loss function | Mamoolan as MSE estefade mishe, az MAE entekhabe behtarie. |
| Optimizer | SGD, Adam, etc. |
